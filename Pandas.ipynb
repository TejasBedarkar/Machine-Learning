# Merge

```python
vr=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
vr1=pd.DataFrame({"A":[1,2,3,4],"C":[31,32,33,34]})
pd.merge(vr,vr1,on="A")#like sql show common part from those table
#how
```

```python
pd.merge(vr,vr1,left_index=True,right_index=True,suffixes=("name","id"))
```

# Concat

```python
sr=pd.Series([1,2,3,4])
sr2=pd.Series([11,21,31,41])
pd.concat([sr,sr2])
```

```python
sr=pd.Series([1,2,3,4])
sr2=pd.Series([11,21,31,41])
pd.concat([sr,sr2],axis=1,keys=["sr","sr2"])
```

```python
o1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
o2=pd.DataFrame({"A":[1,2,3,4],"C":[31,32,33,34]})
pd.concat([o1,o2],axis=1,keys=["o1","o2"])
```

# Join

```python
r=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
r1=pd.DataFrame({"C":[10,20,30,40],"D":[11,22,33,44]},index=["a","b","c","d"])
r.join(r1)
```

```python
t=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
t1=pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])
t1.join(t)
print()
```

```python
t1.join(t,how="right")
```

# Group By

```python
# Data arrange in sequence 
```

```python
f=pd.DataFrame({"name":["a","b","c","d","a","a","b","a","c","b"],"s1":[4,5,6,7,3,21,5,6,9,12],
                "s2":[5,6,7,3,1,2,8,11,2,8]})
f
```

```python
f1=f.groupby("name")
f1
```

```python
for x,y in f1:
    print(x)
    
    print(y)
    print()
```

```python
#want a data of single element
f1.get_group("a")
```

```python
#to make min,max,mean from all data table
f1.min()
```

# Pivot()

```python
h=pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","c","a","b","c"],"eng":[11,12,11,13,14,12],"math":[12,14,11,13,15,14]})
h
```

```python
h.pivot(index="days",columns="st_name")
```

```python
h.pivot(index="days",columns="st_name",values="eng")
#to get a perticular values data
```

```python
h.pivot_table(index="st_name",columns="days",aggfunc="mean")#to mathamathical functions use
```

```python
h.pivot_table(index="st_name",columns="days",aggfunc="sum",margins=True)#margin mens show avg value
```
