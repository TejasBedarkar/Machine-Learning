
# SVM

```python
# 1.Linear SVM
It is used when dataset can be classified into 2 classes using a strait line.
# 2.Non-Linear SVM
It is used when the dataset cannot be classified into 2 classes using a straight line.
```

```python
# KERNEL FUNCTION
It takes low dimensional input space and transfer it into a higher-dimensional space,i.e.,it converts not 
separable problem to separable problem.
```

```python
import pandas as pd
```

```python
df=pd.read_csv("D:\\Rushis Internship\\teja\\IRIS1.csv")
```

```python
;df.head()
```

```python
x=df.iloc[:,0:4]
```

```python
x
```

```python
y=df.iloc[:,4]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel="linear")
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
x_test
```

```python
model.score(x_test,y_test)
```
